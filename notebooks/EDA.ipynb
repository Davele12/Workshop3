# Imports

```python
import pandas as pd
```

# Transformations
For the model we need an specific format.

```python
archivos = ['../features/2015.csv', '../features/2016.csv', '../features/2017.csv', '../features/2018.csv', '../features/2019.csv']
dataframes = []

for archivo in archivos:
    df = pd.read_csv(archivo)
    df['Year'] = archivo.split('/')[-1].split('.')[0]
    dataframes.append(df)

combined_df = pd.concat(dataframes, ignore_index=True)
combined_df.head()
```

```python
print(combined_df.columns)
```

```python
def unir_valores(series):
    valores_no_nulos = series.dropna().unique()
    if len(valores_no_nulos) == 1:
        return valores_no_nulos[0] 
    return ', '.join(map(str, valores_no_nulos))

combined_df['country'] = combined_df[['Country', 'Country or region']].agg(unir_valores, axis=1)
combined_df['happiness_rank'] = combined_df[['Happiness Rank', 'Happiness.Rank', 'Overall rank']].agg(unir_valores, axis=1)
combined_df['happiness_score'] = combined_df[['Happiness Score', 'Happiness.Score', 'Score']].agg(unir_valores, axis=1)
combined_df['GDP'] = combined_df[['Economy (GDP per Capita)', 'Economy..GDP.per.Capita.', 'GDP per capita']].agg(unir_valores, axis=1)
combined_df['life_expectancy'] = combined_df[['Health (Life Expectancy)', 'Health..Life.Expectancy.', 'Healthy life expectancy']].agg(unir_valores, axis=1)
combined_df['freedom'] = combined_df[['Freedom', 'Freedom to make life choices']].agg(unir_valores, axis=1)
combined_df['generosity'] = combined_df[['Generosity']].agg(unir_valores, axis=1)
combined_df['government_corruption'] = combined_df[['Trust (Government Corruption)', 'Trust..Government.Corruption.', 'Perceptions of corruption']].agg(unir_valores, axis=1)
combined_df['dystopia_residual'] = combined_df[['Dystopia Residual', 'Dystopia.Residual']].agg(unir_valores, axis=1)
combined_df['family'] = combined_df[['Family', 'Social support']].agg(unir_valores, axis=1)

columns_to_drop = ['Country or region', 'Happiness.Rank', 'Overall rank', 'Happiness.Score', 'Score',
                   'Economy (GDP per Capita)', 'Economy..GDP.per.Capita.', 'GDP per capita',
                   'Health (Life Expectancy)', 'Health..Life.Expectancy.', 'Healthy life expectancy',
                   'Freedom to make life choices', 'Trust (Government Corruption)', 'Trust..Government.Corruption.',
                   'Perceptions of corruption', 'Dystopia.Residual', 'Social support']

combined_df.drop(columns=columns_to_drop, inplace=True)

combined_df.head()
```
